```python
with open('input.txt') as my_input:
    
    seeds = []
    s_to_s = False
    s_to_f = False
    f_to_w = False
    w_to_l = False
    l_to_t = False
    t_to_h = False
    h_to_l = False
    
    s2s = {}
    s2f = {}
    f2w = {}
    w2l = {}
    l2t = {}
    t2h = {}
    h2l = {}
    
    counter = 0
    
    for line in my_input:
        
        if line.find('seeds') != -1:
            seeds = list(map(int, line.split(':')[1].strip().split()))
            
        if line.find('seed-to-soil') != -1:
            s_to_s = True
            continue
        
        if line.find('soil-to-fertilizer') != -1:
            s_to_s = False
            s_to_f = True
            counter = 0
            continue
            
        if line.find('fertilizer-to-water') != -1:
            s_to_f = False
            f_to_w = True
            counter = 0
            continue
            
        if line.find('water-to-light') != -1:
            f_to_w = False
            w_to_l = True
            counter = 0
            continue
            
        if line.find('light-to-temperature') != -1:
            w_to_l = False
            l_to_t = True
            counter = 0
            continue
            
        if line.find('temperature-to-humidity') != -1:
            l_to_t = False
            t_to_h = True
            counter = 0
            continue
            
        if line.find('humidity-to-location') != -1:
            t_to_h = False
            h_to_l = True
            counter = 0
            continue
        
        if s_to_s and len(line) > 5:
            destination, source, length = line.split()
            destination = int(destination)
            source = int(source)
            length = int(length)
            s2s[counter] = [source, destination, length]
            counter += 1
            
        if s_to_f and len(line) > 5:
            destination, source, length = line.split()
            destination = int(destination)
            source = int(source)
            length = int(length)
            s2f[counter] = [source, destination, length]
            counter += 1
            
        if f_to_w and len(line) > 5:
            destination, source, length = line.split()
            destination = int(destination)
            source = int(source)
            length = int(length)
            f2w[counter] = [source, destination, length]
            counter += 1
            
        if w_to_l and len(line) > 5:
            destination, source, length = line.split()
            destination = int(destination)
            source = int(source)
            length = int(length)
            w2l[counter] = [source, destination, length]
            counter += 1
            
        if l_to_t and len(line) > 5:
            destination, source, length = line.split()
            destination = int(destination)
            source = int(source)
            length = int(length)
            l2t[counter] = [source, destination, length]
            counter += 1
            
        if t_to_h and len(line) > 5:
            destination, source, length = line.split()
            destination = int(destination)
            source = int(source)
            length = int(length)
            t2h[counter] = [source, destination, length]
            counter += 1
            
        if h_to_l and len(line) > 5:
            destination, source, length = line.split()
            destination = int(destination)
            source = int(source)
            length = int(length)
            h2l[counter] = [source, destination, length]
            counter += 1
```

```python
min_loc = 100000000000000
min_seed = 0

for seed in seeds:
    
    soil = 0
    fert = 0
    water = 0
    light = 0
    temp = 0
    humid = 0
    loc = 0
    
    seed_found = False
    soild_found = False
    fert_found = False
    water_found = False
    light_found = False
    temp_found = False
    humid_found = False
    loc_found = False
    
    for mapping in s2s:
        source, destination, length = s2s[mapping]
        if seed >= source and seed < source + length:
            soil = seed - source + destination
            seed_found = True
    
    if not seed_found:
        soil = seed
        
    for mapping in s2f:
        source, destination, length = s2f[mapping]
        if soil >= source and soil < source + length:
            fert = soil - source + destination
            soil_found = True
    
    if not soil_found:
        fert = soil
        
    for mapping in f2w:
        source, destination, length = f2w[mapping]
        if fert >= source and fert < source + length:
            water = fert - source + destination
            fert_found = True
    
    if not fert_found:
        water = fert
        
    for mapping in w2l:
        source, destination, length = w2l[mapping]
        if water >= source and water < source + length:
            light = water - source + destination
            water_found = True
    
    if not water_found:
        light = water
        
    for mapping in l2t:
        source, destination, length = l2t[mapping]
        if light >= source and light < source + length:
            temp = light - source + destination
            light_found = True
    
    if not light_found:
        temp = light
        
    for mapping in t2h:
        source, destination, length = t2h[mapping]
        if temp >= source and temp < source + length:
            humid = temp - source + destination
            temp_found = True
    
    if not temp_found:
        humid = temp
        
    for mapping in h2l:
        source, destination, length = h2l[mapping]
        if humid >= source and humid < source + length:
            loc = humid - source + destination
            humid_found = True
    
    if not humid_found:
        loc = humid
        
    if loc < min_loc:
        min_loc = loc
```

```python
with open("input.txt") as fin:
    lines = fin.read().strip().split("\n")

raw_seeds = list(map(int, lines[0].split(" ")[1:]))
seeds = [
    (raw_seeds[i], raw_seeds[i+1])
    for i in range(0, len(raw_seeds), 2)
]

# Generate all the mappings
maps = []

i = 2
while i < len(lines):
    catA, _, catB = lines[i].split(" ")[0].split("-")
    maps.append([])

    i += 1
    while i < len(lines) and not lines[i] == "":
        dstStart, srcStart, rangeLen = map(int, lines[i].split())
        maps[-1].append((dstStart, srcStart, rangeLen))
        i += 1

    maps[-1].sort(key=lambda x: x[1])

    i += 1


# Ensure that all mappings are disjoint
for m in maps:
    for i in range(len(m)-1):
        if not m[i][1] + m[i][2] <= m[i+1][1]:
            print(m[i], m[i+1])


def remap(lo, hi, m):
    # Remap an interval (lo,hi) to a set of intervals m
    ans = []
    for dst, src, R in m:
        end = src + R - 1
        D = dst - src  # How much is this range shifted

        if not (end < lo or src > hi):
            ans.append((max(src, lo), min(end, hi), D))

    for i, interval in enumerate(ans):
        l, r, D = interval
        yield (l + D, r + D)

        if i < len(ans) - 1 and ans[i+1][0] > r + 1:
            yield (r + 1, ans[i+1][0] - 1)

    # End and start ranges can use some love
    if len(ans) == 0:
        yield (lo, hi)
        return

    if ans[0][0] != lo:
        yield (lo, ans[0][0] - 1)
    if ans[-1][1] != hi:
        yield (ans[-1][1] + 1, hi)


locs = []

ans = 1 << 60

for start, R in seeds:
    cur_intervals = [(start, start + R - 1)]
    new_intervals = []

    for m in maps:
        for lo, hi in cur_intervals:
            for new_interval in remap(lo, hi, m):
                new_intervals.append(new_interval)

        cur_intervals, new_intervals = new_intervals, []

    for lo, hi in cur_intervals:
        ans = min(ans, lo)


print(ans)
```
